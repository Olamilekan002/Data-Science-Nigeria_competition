```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
client_train = pd.read_csv('client_train.csv')
invoice_train = pd.read_csv('invoice_train.csv')
client_test = pd.read_csv('client_test.csv')
invoice_test = pd.read_csv('invoice_test.csv')
```

```python
client_train.head()
```

```python
invoice_train.head()
```

```python
client_train['creation_date'] = pd.to_datetime(client_train['creation_date'])
client_test['creation_date'] = pd.to_datetime(client_test['creation_date'])

client_train['cday'] = client_train['creation_date'].apply(lambda x: x.day)
client_train['cmonth'] = client_train['creation_date'].apply(lambda x: x.month)
client_train['cyear'] = client_train['creation_date'].apply(lambda x: x.year)

client_test['cday'] = client_test['creation_date'].apply(lambda x: x.day)
client_test['cmonth'] = client_test['creation_date'].apply(lambda x: x.month)
client_test['cyear'] = client_test['creation_date'].apply(lambda x: x.year)
```

```python
client_train.drop('creation_date',axis=1,inplace=True)
client_test.drop('creation_date',axis=1,inplace=True)
```

```python
invoice_train.head()
```

```python
res = {'A':10}
invoice_train['counter_statue'] = invoice_train['counter_statue'].replace(res)
```

```python
invoice_train.counter_type.value_counts()
```

```python
invoice_train['counter_statue'] = invoice_train['counter_statue'].astype('int')
invoice_test['counter_statue'] = invoice_test['counter_statue'].astype('int')
```

```python
des = {'ELEC':1,'GAZ':2}
invoice_train['counter_type'] = invoice_train['counter_type'].map(des)
invoice_test['counter_type'] = invoice_test['counter_type'].map(des)
```

```python
invoice_train['tconsumption'] = invoice_train['consommation_level_1'] + invoice_train['consommation_level_2'] + invoice_train['consommation_level_3'] + invoice_train['consommation_level_4']

invoice_test['tconsumption'] = invoice_test['consommation_level_1'] + invoice_test['consommation_level_2'] + invoice_test['consommation_level_3'] + invoice_test['consommation_level_4']
```

```python
invoice_train.drop(['consommation_level_1','consommation_level_2','consommation_level_3','consommation_level_4'],
                   axis=1,inplace=True)

invoice_test.drop(['consommation_level_1','consommation_level_2','consommation_level_3','consommation_level_4'],
                   axis=1,inplace=True)
```

```python
invoice_train.head()
```

```python
invoice_train.dtypes
```

```python
invoice_train['invoice_date'] = pd.to_datetime(invoice_train['invoice_date'])
invoice_test['invoice_date'] = pd.to_datetime(invoice_test['invoice_date'])

invoice_train['iday'] = invoice_train['invoice_date'].apply(lambda x: x.day)
invoice_train['imonth'] = invoice_train['invoice_date'].apply(lambda x: x.month)
invoice_train['iyear'] = invoice_train['invoice_date'].apply(lambda x: x.year)

invoice_test['iday'] = invoice_test['invoice_date'].apply(lambda x: x.day)
invoice_test['imonth'] = invoice_test['invoice_date'].apply(lambda x: x.month)
invoice_test['iyear'] = invoice_test['invoice_date'].apply(lambda x: x.year)
```

```python
invoice_train.drop('invoice_date',axis=1,inplace=True)
invoice_test.drop('invoice_date',axis=1,inplace=True)
```

```python
invoice_train.head()
```

```python
invoice_train.dtypes
```

```python
ninv_train = invoice_train.groupby('client_id').mean()
```

```python
ninv_test = invoice_test.groupby('client_id').mean()
```

```python
ninv_train = ninv_train.reset_index()
ninv_test = ninv_test.reset_index()
```

```python
ninv_train.head()
```

```python
ninv_train.info()
```

```python
invoice_train.info()
```

```python
candi_train = pd.merge(ninv_train,client_train,on='client_id',how='outer')
candi_test = pd.merge(ninv_test,client_test,on='client_id',how='outer')
```

```python
candi_train.head()
```

```python
candi_train['value_day'] = abs(candi_train['iday']-candi_train['cday'])
candi_train['value_month'] = abs(candi_train['imonth']-candi_train['cmonth'])
candi_train['value_year'] = abs(candi_train['iyear']-candi_train['cyear'])

candi_test['value_day'] = abs(candi_test['iday']-candi_test['cday'])
candi_test['value_month'] = abs(candi_test['imonth']-candi_test['cmonth'])
candi_test['value_year'] = abs(candi_test['iyear']-candi_test['cyear'])
```

```python
candi_train['tcoef'] = candi_train['tarif_type']/candi_train['counter_coefficient']
candi_train['tread'] = candi_train['tarif_type']/candi_train['reading_remarque']
candi_train['ttype'] = candi_train['tarif_type']/candi_train['counter_type']
candi_train['concli'] = candi_train['tconsumption']/candi_train['client_catg']
candi_train['tcli'] = candi_train['tarif_type']/candi_train['client_catg']
candi_train['indiff'] = abs(candi_train['new_index'] - candi_train['old_index'])

candi_test['tcoef'] = candi_test['tarif_type']/candi_test['counter_coefficient']
candi_test['tread'] = candi_test['tarif_type']/candi_test['reading_remarque']
candi_test['ttype'] = candi_test['tarif_type']/candi_test['counter_type']
candi_test['concli'] = candi_test['tconsumption']/candi_test['client_catg']
candi_test['tcli'] = candi_test['tarif_type']/candi_test['client_catg']
candi_test['indiff'] = abs(candi_train['new_index'] - candi_test['old_index'])
```

```python
X = candi_train.drop(['target','client_id','counter_statue','months_number','counter_coefficient','cday','cmonth','value_day','value_month','tcoef','indiff'],axis=1)
y = candi_train['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=1000,max_depth=None,min_samples_split=10,max_leaf_nodes=20)
```

```python
rfc.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
rfcpred = rfc.predict(X_test)
rfcprob = rfc.predict_proba(X_test)[:,1]
```

```python
error = 1 - accuracy_score(y_test,rfcpred)
error
```
